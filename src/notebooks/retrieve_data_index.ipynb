```python
from tqdm import tqdm
import pandas as pd
```

```python
import os
import sys
sys.path.append('..')
```

```python
from script_utils import load_dataset
from evaluation.utils_eval import build_pdb_dict
```

```python
root = '../../data/cleaned_crossdocked_data'
cls_name = 'CombinedSparseGraphDataset'
```

```python
train_dataset = load_dataset(cls_name, root, 'train')
valid_dataset = load_dataset(cls_name, root, 'valid')
test_dataset = load_dataset(cls_name, root, 'test')
```

```python
train_dataset[0].ligand_name
```

```python
raw_data_path = '../../data/cleaned_crossdocked_data/raw'
pdb_dict, pdb_rev_dict = build_pdb_dict(raw_data_path)
```

```python
pdb_rev_dict['1zpl_A_rec_1zpl_mag_lig_tt_docked_2.sdf']
```

```python
def retrieve_index(dataset, pdb_rev_dict=pdb_rev_dict):
    index_l = []
    for data in tqdm(dataset):
        lig = data.ligand_name
        folder = pdb_rev_dict[lig + '.sdf']
        if folder not in index_l:
            index_l.append(folder)

    return index_l
```

```python
train_ids = retrieve_index(train_dataset)
valid_ids = retrieve_index(valid_dataset)
test_ids = retrieve_index(test_dataset)
```

```python
test_ids
```

```python
train_id_df = pd.DataFrame(train_ids, columns=['folders'])
valid_id_df = pd.DataFrame(valid_ids, columns=['folders'])
test_id_df = pd.DataFrame(test_ids, columns=['folders'])
```

```python
save_path = train_dataset.processed_dir
save_path
```

```python
train_id_df.to_csv(os.path.join(save_path, 'train_index.csv'), index=False)
valid_id_df.to_csv(os.path.join(save_path, 'valid_index.csv'), index=False)
test_id_df.to_csv(os.path.join(save_path, 'test_index.csv'), index=False)
```
